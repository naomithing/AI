### Notes on Concatenation and Stack.
___
Concatenation:

A concatenation operation joins a sequence of arrays along an *existing* axis. All arrays must either have the same shape (except in the concatenating dimension) or be empty.

Stacking:

The stack operation joins a sequence of arrays along a *new* axis. The axis parameter specifies the index of the new axis in the dimensions of the result. For example, if axis=0 it will be the first dimension and if axis=-1 it will be the last dimension. All arrays need to be of the same size.  The stacked array has one more dimension than the input arrays.

```python
# Concatenating Numpy Arrays

array1 = np.random.randint(3, size = (3, 4, 5))
print("Array 1 is \n", array1, " with dimensions ", array1.shape, "\n")

array2 = np.random.randint(4, size = (3, 4, 5))
print("Array 2 is \n", array2, " with dimensions ", array2.shape, "\n")

concatenated_array1 = np.concatenate((array1, array2), axis = 0)
print("Concatenated array 1 is \n", concatenated_array1, "\n\n", "and the dimensions of the concatenated array 1 are: \n", concatenated_array1.shape)

concatenated_array2 = np.concatenate((array1, array2), axis = 1)
print("Concatenated array 2 is \n", concatenated_array2, "\n\n", "and the dimensions of the concatenated array 2 are: \n", concatenated_array2.shape)

concatenated_array3 = np.concatenate((array1, array2), axis = 2)
print("Concatenated array 3 is \n", concatenated_array3, "\n\n", "and the dimensions of the concatenated array 3 are: \n", concatenated_array3.shape)
```

Concatenation Observation:
As per observation, using random integers, this code creates two 3-dimensional arrays, {array1} and {array2}. Next, it uses 'np.concatenate()' to demonstrate various concatenation methods.The arrays are vertically stacked along axis 0, which increases the size along hte first dimension. The arrays are aligned side by side when concatenated along axis 1, which increases the size along the second dimension. Lastly, the arrays are extended along the third dimension but concatenation along axis 2. The final concatenated arrays and their sizes are shown, giving a clear picture of the several arrangements depending on the selected concatenation axis.

```python
# Stacking Numpy Arrays

array1 = np.random.randint(3, size = (3, 4, 5))
print("Array 1 has dimensions ", array1.shape, "\n")

array2 = np.random.randint(4, size = (3, 4, 5))
print("Array 2 has dimensions ", array2.shape, "\n")

stacked_array1 = np.stack((array1, array2), axis = 0)
print("Stacked array 1 has dimensions", stacked_array1.shape, "\n")

stacked_array2 = np.stack((array1, array2), axis = 1)
print("Stacked array 2 has dimensions", stacked_array2.shape, "\n")

stacked_array3 = np.stack((array1, array2), axis = 2)
print("Stacked array 3 has dimensions", stacked_array3.shape, "\n")

stacked_array4 = np.stack((array1, array2), axis = -1)
print("Stacked array 4 has dimensions", stacked_array4.shape, "\n")
```

Stacking Observation:
As per the observation, this code generated 'array1' and 'array2' with the help of random integers. It then shows how to stack them in different ways using 'np.stack()'. Think of it arranging books on a shelf: stacking along axis 0 adds more books to the top shelf, axis 1 lines them up side by side, and axis 2 stacks them along the third dimension. The code prints the new arrangements, giving a quick peek at how stacking changes the way the numbers are organized.

```python
import math

def euclidean_norm(vector):
    sum = 0
    for x in vector:
      sum= sum+x**2
      norm = math.sqrt(sum)
    return norm

import numpy as np

my_vector = [0.5, -1.2, 3.3, 4.5]
result = euclidean_norm(my_vector)
print(result)
```

```python
import numpy as np

def angle_between_two_vectors(vector1, vector2):
    value = np.dot(vector1, vector2)
    vec_value = np.dot(np.linalg.norm(vector1), np.linalg.norm(vector2))
    angle_degrees = np.degrees(np.arccos(value/vec_value))
    return angle_degrees

# Example usage:
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])
angle = angle_between_two_vectors(vector1, vector2)
print(f"Angle between vector1 and vector2: {angle} degrees")
```

### Task set VI: Solve system fo linear equations:
___
Solve the following system of Linear Equations using inverse method and numpy.

\begin{align}
2x - 3y + z = -1\\
x - y +2z = -3\\
3x + y -z = 9
\end{align}


```python
import numpy as np
A = np.array([[2,-3,1],[1,-1,2],[3,1,-1]])
B = np.array([-1,-3,9])
inverse_of_A = np.linalg.inv(A)
result = np.matmul(inverse_of_A, B)
print("The answer: ", result)
```

The system of linear equations can also be solved without explicitly computing the inverse;  by using the numpy.linalg.solve function. This method is numerically more stable and efficient compared to computing the inverse directly. Here's how you can do it:

```python
import numpy as np
A = np.array([[2,-3,1],[1,-1,2],[3,1,-1]])
B = np.array([-1,-3,9])
result = np.linalg.solve(A,B)
print(result)
```
