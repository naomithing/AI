# Data Cleaning and Summary Statistics

```python
# Replace the null values with 0.15 values
winedt['citric acid'].fillna(winedt['citric acid'].mean(), inplace = True)
# Replace the null values with mean values
winedt['chlorides'].fillna(winedt['chlorides'].mean(), inplace = True)
winedt['sulphates'].fillna(winedt['sulphates'].mean(), inplace = True)
winedt['alcohol'].fillna(winedt['alcohol'].mean(), inplace = True)
print("After cleaning the data:")
print(winedt.isnull().sum())
```

The reason I chose the `fillna()` method to address missing values is its particular suitability for numeric values. For the first missing column, I manually inputted a value, while for the other missing columns, I utilized the mean value. This approach ensures that the overall statistical properties of the column are maintained while addressing the missing values.

# Visualize the Data: Make Interpret and Save your Charts

# Univariate Analysis


```python
plt.figure(figsize=(15,10))
sns.set(style="whitegrid")
line_chart = sns.countplot(x=ndt['quality'],palette="GnBu")
line_chart.bar_label(line_chart.containers[0])
plt.xlabel('quality')
plt.ylabel('count')
plt.title('Showing the quality of wine')
plt.savefig('bar.png')
plt.show()
```

```python
volaci_skewness = ndt['volatile acidity'].skew()
plt.figure(figsize=(15,10))
sns.histplot(ndt['volatile acidity'],kde = True, bins=5,color='turquoise',edgecolor='black')
sns.set(style="whitegrid")
plt.title("Distribution of volatile acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel("Number of observation")
plt.savefig("Hist.png")
plt.show()
```

```python
sns.boxplot(ndt['volatile acidity'])
plt.title("Outliers of volatile acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel("Values")
plt.legend(['volatile acidity'])
plt.savefig("VA.png")
plt.show
```

# Bi-variate Analysis

```python
col_num = ['citric acid','chlorides']
sns.pairplot(ndt[col_num], height = 4)
plt.suptitle("Pairplot of citric acid and chlorides", y=1.05)
plt.savefig("PP.png")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("Correlation of 'Wine Quality Dataset'")
sns.heatmap(ndt.corr(), annot=True,cmap="Reds")
plt.savefig("Correlation.png")
plt.show()
```
